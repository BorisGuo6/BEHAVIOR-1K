```python
import json

with open(r"D:\ig_pipeline\artifacts\pipeline\object_inventory.json") as f:
    i = json.load(f)
```

```python
import csv

renames = {}
with open(r"D:\ig_pipeline\metadata\object_renames.csv") as f:
    for r in csv.DictReader(f):
        renames[r["Object name"]] = r["New Category"]
```

```python
renamed_providers = {}
for k, v in i["providers"].items():
    cat, mod = k.split("-")
    if k in renames:
        cat = renames[k]
    renamed_providers[f"{cat}-{mod}"] = v
```
