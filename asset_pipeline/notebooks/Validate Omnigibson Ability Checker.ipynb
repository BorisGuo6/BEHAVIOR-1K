```python
tro_synsets = {s for s in Synset.all_objects() if len(s.tasks) > 0 and s.state != STATE_SUBSTANCE}
```

```python
from bddl.object_taxonomy import ObjectTaxonomy
OT = ObjectTaxonomy()
```

```python
no_good_obj = {
  s for s in tro_synsets
  if all(
    len(get_all_object_category_models_with_abilities(c.name, OT.get_abilities(s.name))) == 0
    for c in {obj.category for obj in s.matching_objects}
  ) 
}
```
