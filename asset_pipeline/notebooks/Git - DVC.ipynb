```python
import os, glob, math
legacy_stuff = [x.replace("D:\\ig_pipeline\\", "") for x in glob.glob(r"D:\ig_pipeline\cad\objects\legacy_*\processed.max")] + [x.replace("D:\\ig_pipeline\\", "") for x in glob.glob(r"D:\ig_pipeline\cad\objects\legacy_*\textures")]

groups = []
groups_of = 100
total = math.ceil(len(legacy_stuff) / groups_of)
for i in range(total):
    f = i * groups_of
    t = (i + 1) * groups_of
    r = legacy_stuff[f:t]
    cmd = "dvc add " + " ".join(r)
    assert len(cmd) < 8000
    groups.append(cmd)
    
print("\n".join(groups))
```
