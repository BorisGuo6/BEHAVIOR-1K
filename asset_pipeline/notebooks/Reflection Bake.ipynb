```python
from PIL import Image
import pathlib
import numpy as np
from tqdm.notebook import tqdm
```

```python
files = list(pathlib.Path(r"D:\BEHAVIOR-1K\asset_pipeline").glob("cad/*/*/bakery/*Reflection*.png"))
print(len(files))
```

```python
# Go through all the files and get all the unique colors in each of them
uniques_by_file = {}
for fn in tqdm(files):
  img = Image.open(fn)
  arr = np.array(img)
  uniques_by_file[fn] = np.unique(arr, return_counts=True)
```

```python
how_many_unique = {k: len(v[0]) for k, v in uniques_by_file.items()}
bad_guys = sorted(how_many_unique.items(), key=lambda x: -x[1])
```
