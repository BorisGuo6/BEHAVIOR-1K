```python
from bddl.knowledge_base import *
```

```python
from collections import defaultdict
group_by_provider = defaultdict(set)
for ps in ParticleSystem.all_objects():
  for particle in ps.particles:
    if "substance" not in particle.provider:
      print(f"Particle {particle.name} in ParticleSystem {ps.name} has non substance provider {particle.provider}")
      group_by_provider[particle.provider].add(particle)
    if "scenes/" in particle.provider:
      print(f"Particle {particle.name} in ParticleSystem {ps.name} has a provider that is a scene, which is not allowed.")
```

```python
for provider, particles in group_by_provider.items():
  print(provider, len(particles))
  for particle in particles:
    print(f"  {particle.name}")
  print()
```
