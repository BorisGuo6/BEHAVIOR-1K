```python
import hashlib, pathlib, collections
from tqdm.notebook import tqdm
```

```python
# Get all the texture files into counters based on filename and hash
ignore_exts = [".txt", ".cube", ".mtl", ".ies"]
texture_counters = collections.defaultdict(lambda: collections.defaultdict(set))
texture_files = [x for x in pathlib.Path(r"D:\ig_pipeline").glob("cad/*/*/textures/*") if not x.suffix in ignore_exts]
for texture_file in tqdm(texture_files):
    # Calculate the hash of the file
    with open(texture_file, "rb") as f:
        file_hash = hashlib.sha256(f.read()).hexdigest()
    
    # Use the filename without extension and the hash as keys
    filename = texture_file.name
    texture_counters[filename][file_hash].add(texture_file)
```

```python
# What are the textures that have multiple hashes?
duplicate_textures = {filename: counter for filename, counter in texture_counters.items() if len(counter) > 1}
```
