```python
from PIL import Image, ImageDraw, ImageFont
import textwrap

# create an image
out = Image.new("RGB", (1000, 500), (255, 255, 255))
# out = Image.open(r"C:\Users\cgokmen\Downloads\splash.png")
HEIGHT = out.height
WIDTH = out.width

# get a font
lines = textwrap.wrap("LEMON-PEPPERINO SEASONING", 12)
assert len(lines) <= 3
text = "\n".join(lines)
FONT_SIZE = 172
fnt = ImageFont.truetype("impact.ttf", FONT_SIZE, encoding="unic")
# get a drawing context
d = ImageDraw.Draw(out)

d.multiline_text((WIDTH // 2, HEIGHT // 2), text, font=fnt, anchor="mm", align="center", fill=(30, 30, 30))
```
