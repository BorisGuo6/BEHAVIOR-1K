```python
import sys
sys.path.append(r"D:\ig_pipeline")

from b1k_pipeline.utils import PIPELINE_ROOT

import json
import csv
```

```python
# Load the categories CSV file
cat_csv = PIPELINE_ROOT / "metadata/category_mapping.csv"
cur_cat_specs = {}
with open(cat_csv, "r") as f:
    csvfile = csv.DictReader(f)
    categories = []
    for row in csvfile:
        cat = row["category"]
        categories.append(cat)
        if row["enable_ag"]:
            cur_cat_specs[cat] = row
```

```python
# Load the avg category specs file
import igibson
import pathlib
avg_cat_specs_file = pathlib.Path(igibson.ig_dataset_path) / "metadata/avg_category_specs.json"
with open(avg_cat_specs_file, "r") as f:
    cat_specs = json.load(f)
```

```python
next(iter(cat_specs.values()))
```

```python
# Print the masses
found_new = []
for cat in categories:
    if cat in cat_specs:
        print(cat_specs[cat]["enable_ag"], *cat_specs[cat]["size"])
    elif cat in cur_cat_specs:
        found_new.append(cat)
        print(int(cur_cat_specs[cat]["enable_ag"]), cur_cat_specs[cat]["size_x"], cur_cat_specs[cat]["size_y"], cur_cat_specs[cat]["size_z"])
    else:
        print("    ")
```

```python
# Load the sampleability file
non_sampleable_cats_file = pathlib.Path(igibson.ig_dataset_path) / "metadata/non_sampleable_categories.txt"
with open(non_sampleable_cats_file, "r") as f:
    non_sampleable_cats = {x.strip() for x in f.readlines()}
```
