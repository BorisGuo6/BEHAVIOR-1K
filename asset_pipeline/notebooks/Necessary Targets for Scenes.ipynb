```python
import json, pathlib

inventory_future = json.loads(pathlib.Path(r"D:\BEHAVIOR-1K\asset_pipeline\artifacts\pipeline\object_inventory.json").read_text())
```

```python
scenes_to_load = [
    "Rs_garden",
    "Wainscott_0_garden",
    "Pomaria_0_garden",
    "Merom_0_garden",
    "Beechwood_0_garden"
]
```

```python
needed_objs = set()
for scene in scenes_to_load:
    tgt = "scenes/" + scene
    for k, vs in inventory_future["needed_by"].items():
        if tgt in vs:
            needed_objs.add(k.split("-")[1])
```

```python
needed_objs
```

```python
needed_providers = set()
for obj, provider in inventory_future["providers"].items():
    if obj.split("-")[1] in needed_objs:
        needed_providers.add(provider)
print("\n".join(sorted(needed_providers)))
```
