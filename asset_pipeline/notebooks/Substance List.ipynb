```python
import csv
with open("/home/cgokmen/Downloads/BEHAVIOR-1K Synsets.csv") as f:
    syn = {r["Name"]: r for r in csv.DictReader(f)}
```

```python
syn
```

```python
substance_needs_asset = {k for k, x in sub.items() if x["type"] != "fluid"}
synset_has_asset = {k for k, v in syn.items() if int(v["Total Objects"]) > 0}
substance_has_asset = {k for k in substance_needs_asset if any(s.split(".")[0] == k for s in synset_has_asset)}
substance_missing_asset = substance_needs_asset - substance_has_asset

print("Needs", len(substance_needs_asset))
print("Has", len(substance_has_asset))
print("Missing", len(substance_missing_asset))
```
