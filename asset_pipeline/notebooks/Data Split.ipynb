```python
import os, glob, shutil, pathlib, random
```

```python
inpath = r"D:\rgb"
outpath = r"D:\sample-dataset-ig2"
os.makedirs(outpath, exist_ok=True)
```

```python
for subdir in os.listdir(inpath):
    fullpath = os.path.join(inpath, subdir)
    if not os.path.isdir(fullpath):
        continue
    
    images = os.listdir(fullpath)
    random.shuffle(images)
    split_at = int(len(images) / 10)
    splits = {
        "val": images[:split_at],
        "train": images[split_at:],
    }
       
    for split_name, split_files in splits.items():
        full_outpath = os.path.join(outpath, split_name, subdir)
        os.makedirs(full_outpath, exist_ok=True)
        
        for f in split_files:
            in_file = os.path.join(fullpath, f)
            out_file = os.path.join(full_outpath, f)
            
            print(in_file, out_file)
            shutil.copy2(in_file, out_file)
```
