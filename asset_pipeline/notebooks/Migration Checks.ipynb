```python
# Check if there are any extra files in ig_cad
for f in to_pipeline:
    items = {x.name for x in f.iterdir()}
    expected = {"processed.max", "processed.max.dvc", "raw.max", "raw.max.dvc", "textures", "textures.dvc", ".gitignore", "artifacts"}
    unexpected = items - expected
    if unexpected:
        print(f"{f} has unexpected entries: {unexpected}")
```
