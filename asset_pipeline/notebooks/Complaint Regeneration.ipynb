```python
import glob, pathlib, json
jsons = {}
complaint_glob = r"D:\ig_pipeline\qa-merged\*\*.json"
for fn in glob.glob(complaint_glob):
    fn = pathlib.Path(fn)
    with open(fn, "r") as f:
        jsons[fn.stem] = json.load(f)["complaints"]
```

```python
with open(r"D:\ig_pipeline\artifacts\pipeline\object_inventory.json", "r") as f:
    providers = {x.split("-")[1]: y for x, y in json.load(f)["providers"].items()}
```

```python
import collections

complaints_by_target = collections.defaultdict(list)
for obj, complaints in jsons.items():
  complaints_by_target[providers[obj]].extend(complaints)
```

```python
complaints_by_target
```

```python
# Save all the complaints
import os
for target, complaints in complaints_by_target.items():
    with open(os.path.join(r"D:\ig_pipeline\cad", target, "complaints.json", "w")) as f:
        json.dump(complaints, f, indent=2)
```
